```python
def custom_len(a):
  n=0
  for i in a:
    n+=1
  return n
```

```python
def custom_max(runs):
  n=len(runs)
  max_runs=runs[0]
  for i in range(1,n):
    if runs[i] > max_runs:
      max_runs=runs[i]
  return max_runs
```

```python
custom_max(runs)
```

```python
def custom_min(runs):
  n=len(runs)
  min_runs=runs[0]
  for i in range(1,n):
    if runs[i] < min_runs:
      min_runs=runs[i]
  return min_runs
```

```python
custom_min(runs)
```

```python
def generate_matrix(rows,colunms):
  matrix=[]
  for i in range(rows):
    row=[]
    for j in range(colunms):
      row.append(int(input()))
    matrix.append(row)
  return matrix
```

```python
generate_matrix(2,2)
```

```python
def sum_all(matrix):
  sum_all=0
  for i in matrix:
    for j in i:
      sum_all+=j
  return sum_all
```

```python
sum_all(matrix)
```

```python
def sum_each_row(matrix):
  row=len(matrix)
  colunm=len(matrix)

  for i in range(row):
    sum_row=0
    for j in range(colunm):
      sum_row +=matrix[i][j]
    print(sum_row)
```

```python
sum_each_row(matrix)
```
